# RSVP Simulation Suite Notebook

This notebook demonstrates the use of `rsvp_simulations.py` for:

1. Phase portraits in $(\Phi, S)$.
2. Bifurcation diagrams vs. $\gamma_{\text{err}}$.
3. 1D PDE simulations and animations.


```python
import numpy as np
import matplotlib.pyplot as plt
from rsvp_simulations import (
    RSVPParams,
    plot_phase_portraits,
    bifurcation_scan,
    run_rsvp_1d_simulation,
    animate_rsvp_1d,
    demo_phi_2d_gaussian,
)

%matplotlib inline
params = RSVPParams()
```

```python
# Phase portraits in (Phi, S)
plot_phase_portraits(params, n_grid=6)
plt.show()
```

```python
# Bifurcation diagram: final Phi vs gamma_err
gamma_vals = np.linspace(0.1, 2.5, 40)
bifurcation_scan(params, gamma_vals)
plt.show()
```

```python
# 1D PDE simulation snapshot
x, t_arr, phi_hist, v_hist, S_hist = run_rsvp_1d_simulation(
    L=10.0, nx=256, t_final=20.0, dt=1e-3, params=params
)
plt.figure(figsize=(7,4))
plt.plot(x, phi_hist[0], label='t=0')
plt.plot(x, phi_hist[-1], label=f't={t_arr[-1]:.1f}')
plt.xlabel('x')
plt.ylabel(r'$\Phi(x)$')
plt.title('1D RSVP Collapse (Initial vs Final)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Optional: 2D Phi demo
X, Y, phi_final = demo_phi_2d_gaussian()
plt.figure(figsize=(5,4))
plt.imshow(phi_final, origin='lower',
           extent=(X.min(), X.max(), Y.min(), Y.max()))
plt.colorbar(label=r'$\Phi$')
plt.title('2D Phi Field after Evolution toward Gaussian S')
plt.xlabel('x')
plt.ylabel('y')
plt.tight_layout()
plt.show()
```

For interactive 1D animations, call `animate_rsvp_1d(...)` in a live Python
session instead of a static notebook, or use `%matplotlib notebook` backend
and display the resulting animation object.
